```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(
    '/Users/rammaruthi/Desktop/AI TRAINING/Datasets/Position_Salaries.csv')
df
```

```python
df.drop('Position', axis=1)
df
```

```python
X = df['Level'].values
y = df['Salary'].values
```

```python
X = np.array([X])
y = np.array([y])
```

```python
X.reshape(-1, 1)
y.reshape(-1, 1)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_features = PolynomialFeatures(degree=2)
X_poly = poly_features.fit_transform(X)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_poly, y)
```

```python
prediction = model.predict(X_poly)
```

```python
plt.scatter(X, prediction)
plt.scatter(X, y)
```
