```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv(
    '/Users/rammaruthi/Desktop/AI TRAINING/Datasets/Housing_Price.csv')
dataset
```

```python
X = dataset.iloc[:, 0:1].values
y = dataset.iloc[:, 1].values
```

```python
X
```

```python
y
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X, y, color='r')
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
pred_values = model.predict(X)
```

```python
plt.scatter(pred_values, y)
plt.plot(X, pred_values)
```

```python
plt.plot(X, pred_values)
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly_features = PolynomialFeatures(degree=3)
X_poly = poly_features.fit_transform(X)
```

```python
X_poly
```

```python
mymodel = LinearRegression()
```

```python
mymodel.fit(X_poly, y)
```

```python
poly_predict = mymodel.predict(X_poly)
```

```python
plt.scatter(X, poly_predict)
plt.scatter(X, y)
plt.plot(X, poly_predict)
```
