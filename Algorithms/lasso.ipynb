```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/rammaruthi/Desktop/AI TRAINING/Datasets/auto-mpg.csv')
df
```

```python
X = df.drop(['mpg','car name'],axis = 1)
y = df[['mpg']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.preprocessing import MinMaxScaler
scale = MinMaxScaler()
```

```python
X_train_transform = scale.fit_transform(X_train)
X_test_transform = scale.fit_transform(X_test)
```

```python
X_train
```

```python
X_train_transform
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train_transform,y_train)
```

```python
y_pred = model.predict(X_test_transform)
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,y_pred)
score
```

```python
from sklearn.linear_model import Lasso
```

```python
lassoModel = Lasso(alpha=0.5)
```

```python
lassoModel.fit(X_train_transform,y_train)
```

```python
y_train_pred = lassoModel.predict(X_train_transform)
y_test_pred = lassoModel.predict(X_test_transform)
```

```python
from sklearn.metrics import r2_score
lassoScore = r2_score(y_test,y_test_pred)
```
