```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/rammaruthi/Desktop/AI TRAINING/Datasets/weather_golf.csv')
```

```python
df
```

```python
outlook_dict = {
  'Sunny' : 0,
  'Rain' : 1,
  'Overcast' : 2
}
```

```python
df['Outlook'] = df['Outlook'].map(outlook_dict)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df['Outlook'] = le.fit_transform(df['Outlook'])
df['Temp'] = le.fit_transform(df['Temp'])
df['Humidity'] = le.fit_transform(df['Humidity'])
df['Wind'] = le.fit_transform(df['Wind'])
```

```python
df
```

```python
X = df.drop('GolfPlayers',axis=1)
y = df['GolfPlayers']
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model = DecisionTreeRegressor()
```

```python
model.fit(X,y)
```

```python
from sklearn import tree
```

```python
fig = plt.figure(figsize=(25,20))
tree.plot_tree(model,feature_names=df.columns,filled=True)
```
